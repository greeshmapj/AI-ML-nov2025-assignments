```python
import pandas as pd

def remove_outliers(df, columns):
    df_filtered = df  # Make a copy of the original DataFrame

    for i in columns:
        Q1 = df[i].quantile(0.25)
        Q3 = df[i].quantile(0.75)
        IQR = Q3 - Q1

        lower_whisker = Q1 - 1.5 * IQR
        upper_whisker = Q3 + 1.5 * IQR

        # Filter out rows where column values are outliers
        df_filtered = df_filtered[(df_filtered[i] >= lower_whisker) & (df_filtered[i] <= upper_whisker)]

    return df_filtered
```

# plan b: we can cap the outliers

```python
import pandas as pd

# Sample DataFrame
data = {
    'A': [10, 12, 14, 120, 16, 18, 20,-88],
}
df = pd.DataFrame(data)

# Specify the column to cap outliers
column = 'A'

# Calculate Q1, Q3, and IQR
Q1 = df[column].quantile(0.25)
Q3 = df[column].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper whiskers
lower_whisker = Q1 - 1.5 * IQR
upper_whisker = Q3 + 1.5 * IQR

#  'A': [10, 12, 14, 120, 16, 18, 20,-88],
```

```python
lower_whisker
```

```python
upper_whisker
```

```python
# Cap outliers using boolean indexing
df.loc[df["A"] < lower_whisker, "A"] = lower_whisker
df.loc[df["A"] > upper_whisker, "A"] = upper_whisker

df

#'A': [10, 12, 14, 120, 16, 18, 20,-88],
```

# method 2: z score

```python

# Load the Titanic dataset
df = sns.load_dataset("titanic")
df
# skewness
```

```python
x=df["fare"]
x
```

```python
mean=df["fare"].mean()
mean
```

```python
sd=df["fare"].std()
sd
```

```python
df["zscore"]=(x-mean)/sd
```

```python
df
```

```python
df2_cleaned = df[(df['zscore'] >= -.5) & (df['zscore'] <= .5)]
df2_cleaned
```
