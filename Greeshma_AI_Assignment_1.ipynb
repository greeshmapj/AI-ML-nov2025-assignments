1. Simple ATM Calculator
Write a program that takes the user's account balance and withdrawal amount. If
sufficient balance is available, deduct the amount and display the updated balance.
Otherwise, print an error message.
Sample Input:
Balance = 5000
Withdrawal = 1500
Sample Output:
Transaction successful. Remaining balance: ₹3500

```python
#1. Simple ATM Calculator

balance = float(input("Please enter your balance: Rs"))
withdrawal = float(input("Please enter your withdrawal amount: Rs"))

if balance>=withdrawal:
  balance = balance - withdrawal
  print("Transaction of Rs",withdrawal,"successful. Remaining balance is Rs",balance)
else:
  print("Oops, insufficient balance")
```

3. Temperature Converter
Write a Python program to convert a given temperature value between Celsius and
Fahrenheit. Accept input as a number and a unit indicator ('C' or 'F').
Sample Input:
Temp = "37"
Unit = "C"
Sample Output:
Temperature in Fahrenheit: 98.6°F

```python
#3. Temperature Converter
temp = float(input("enter the temperature: "))
unit = input("enter unit as 'C' or 'F': ").strip().upper()

if unit == 'C':
  temp = (temp*9/5) + 32
  print(f"Temperature in Fahrenheit is {temp}F")
elif unit == 'F':
  temp = (temp-32)*5/9
  print(f"Temperature in Celsius is {temp}C")
else:
  print("Invalid Input, Try again with C or F")
```

4. Mobile Recharge Validity
Write a program that takes the recharge date and today’s date as input and checks
whether the 28-day recharge plan is still valid or expired.
Sample Input:
Recharge date: 2025-07-01
Today: 2025-07-30
Sample Output:
Plan expired. Recharge again.

```python
#4. Mobile Recharge Validity
from datetime import datetime
#inputs
recharge_date = input("Enter recharge date (YYYY-MM-DD): ")
today = input("Enter today's date (YYYY-MM-DD): ")

#str to datetime convvertion for calculation
recharge = datetime.strptime(recharge_date, "%Y-%m-%d")
today_date = datetime.strptime(today, "%Y-%m-%d")

days_passed = (today_date - recharge).days

if days_passed > 28:
    print("Plan expired. Recharge again!")
else:
    print("Plan valid")

```

5. Bus Seat Reservation System:
A list contains the names of passengers in order of booking. A cancellation occurs; remove
the name from the list and shift the rest. Also, insert a new booking at a specified position.
Sample Input:
Booked = ['Amit', 'Sara', 'John'],
Cancel = 'Sara',
New = 'Neha', Position = 1
Sample Output:
['Amit', 'Neha', 'John']

```python
#5. Bus Seat Reservation System:

booked = input("Enter all booked passengers name(comma separated): ").split(',')
# Remove extra spaces
booked = [name.strip() for name in booked]
cancelled = input("enter the cancelled passengers name: ")

if cancelled in booked:
    booked.remove(cancelled)
else:
    print("Passenger not found in booking list")

#new booking
new_booking = input("enter the passenger to be added: ")
insert_position = int(input("enter the position to which the passenger need to be added: "))
if 0 <= insert_position <= len(booked):
    booked.insert(insert_position, new_booking)
else:
    print("Invalid position")

# Final output
print("Updated Booking List:", booked)
```

6. Sum of even numbers
Write a program that takes a number as input and prints the sum of all even numbers
from 1 to that number.
Sample Input:
Number: 10
Sample Output:
Sum of even numbers: 30
(2 + 4 + 6 + 8 + 10)bold text

```python
#6. Sum of even numbers in given range
number = int(input("Enter a number: "))
even_numbers = []
total = 0
for i in range(1,number+1):
  if i%2==0:
    even_numbers.append(i)
    total+=i

calculation = " + ".join(map(str, even_numbers))
print(f"sum of even numbers: {total}({calculation})")
```

7. Dictionary Student Marks Summary
Write a program that takes input of 3 students’ names and their marks, stores them in a
dictionary, and prints the average marks.
Sample Input:
Enter name: Alice   
Enter marks: 80   
Enter name: Bob   
Enter marks: 90   
Enter name: Charlie   
Enter marks: 70
Sample Output:
Average Marks: 80.0

```python
#7. Dictionary – Student Marks Summary
students = {}
for i in range(3):
    name = input("Enter name: ")
    marks = int(input("Enter marks: "))
    students[name] = marks

average = sum(students.values()) / len(students)
print("Average Marks:", average)
```

8. Dictionary – Word Frequency
Write a program that takes a sentence and displays the frequency of each word using a
dictionary.
Sample Input:
Sentence: hello world hello python
Sample Output:
hello: 2   
world: 1   
python: 1

```python
#8. Dictionary – Word Frequency
count=0
freq={}
sent = input("enter the sentence: ")
new_word = sent.split()
for words in new_word:
  if words in freq:
      freq[words] += 1
  else:
      freq[words] = 1

for words, count in freq.items():
    print(f"{words}: {count}")
```

9. Functions – Check Prime
Write a function that checks whether a number is prime. Call the function with user input.
Sample Input:
Enter a number  : 17
Sample output:
17 is a prime number.

```python
#9. Functions – Check Prime
number = int(input("Enter a number: "))

def is_prime(number):
  if number>1:
    for i in range(2,number):
      if number%i==0:
        return False
    return True
  else:
    return False

if is_prime(number):
  print(f"{number} is a prime number")
else:
  print(f"{number} is not a prime number")

```

10. Vowel Counter  
Write a function that takes a string and returns the number of vowels in it.  
Sample Input:
Enter a string: Python Programming
Sample output:
Number of vowels: 4

```python
#10. Vowel Counter

def vowel_count(sent):
    vowels = ['a', 'e', 'i', 'o', 'u']
    count = 0

    for char in sent.lower():
        if char in vowels:
            count += 1

    return count

sent = input("Enter a string: ")
result = vowel_count(sent)
print("Number of vowels:", result)
```

11. Count Divisible Numbers
Write a program that takes two numbers as input: a range limit N and a divisor D. Count
how many numbers between 1 and N are divisible by D.
Sample Input:
Enter the range limit (N): 50   
Enter the divisor (D): 5
Sample output:
Count of numbers divisible by 5: 10
(i.e., 5, 10, 15, ..., 50)

```python
#11. Count Divisible Numbers
range_limit = int(input("Enter the range limit(N): "))
divisor = int(input("Enter the divisor(D): "))
count = 0
divisible = []

for i in range(1,range_limit+1):
  if i%divisor==0:
    count+=1
    divisible.append(i)

print(f"Count of numbers divisible by {divisor}:{count}")
print(divisible)
```
