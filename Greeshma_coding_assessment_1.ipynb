# **1. Load Dataset**



```python
import pandas as pd
test_df = pd.read_csv('/content/test_2umaH9m.csv')
train_df = pd.read_csv('/content/train_LZdllcl.csv')
sample_df = pd.read_csv('/content/sample_submission_M0L0uXE.csv')
train_df.head(3)
```

```python
test_df.head(3)
```

```python
train_df.info()
```

**INSIGHT**

* Target : is_promoted

```python
#Handling duplicates
train_df.duplicated().sum()
```

No duplicates present

# **2. Missing Values Handling**

```python
#Handling Missing Values
round(train_df.isnull().mean()*100,2)
```

```python
test_df.info()
```

```python
train_df['education'].fillna(train_df['education'].mode()[0],inplace=True)
test_df['education'].fillna(test_df['education'].mode()[0],inplace=True)

train_df['previous_year_rating'].fillna(train_df['previous_year_rating'].median(), inplace=True)
test_df['previous_year_rating'].fillna(train_df['previous_year_rating'].median(),inplace=True)
```

```python
train_df.info()
```

# **3. Feature Selection**


```python
train_df.drop('employee_id', axis=1, inplace=True)
test_df.drop('employee_id', axis=1, inplace=True)
```

```python
#Checking class imbalance
train_df['is_promoted'].value_counts(normalize=True)
```

Class imbalance exists, will focus on F1 score

# **5.Outlier Handling**




```python
train_df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
for col in train_df.columns:
  sns.boxplot(train_df[col])
  plt.show()
```

```python

skew_check=train_df.select_dtypes(include=['number'])
skew_check.skew()
```

```python
#log transformation for outlier handling
import numpy as np

cols_to_transform = ['no_of_trainings', 'length_of_service', 'age']

for col in cols_to_transform:
    train_df[col] = np.log1p(train_df[col])
    test_df[col] = np.log1p(test_df[col])
```

# **6. Encoding**

```python
from sklearn.preprocessing import LabelEncoder

cat_cols = train_df.select_dtypes(include='object').columns
print(cat_cols)

le = LabelEncoder()
for col in cat_cols:
    train_df[col] = le.fit_transform(train_df[col])
    test_df[col] = le.transform(test_df[col])
```

```python
train_df.head(3)
```

# **7. Split Features & Target**

```python
X = train_df.drop('is_promoted', axis=1)
y = train_df['is_promoted']
```

# **8. Train-test split**

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y,test_size=0.2,stratify=y,random_state=42)
```

# **9. Model Building**

```python
#Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(class_weight='balanced')
model.fit(X_train, y_train)
```

# **10. Evaluate Model**

```python
from sklearn.metrics import f1_score

y_pred = model.predict(X_val)
print(f1_score(y_val, y_pred))
```

# **11. Hyperparameter Tuning**

```python
from sklearn.model_selection import RandomizedSearchCV

param_dist = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

rf = RandomForestClassifier(class_weight='balanced', random_state=42)

random_search = RandomizedSearchCV(
    rf,
    param_distributions=param_dist,
    n_iter=10,
    scoring='f1',
    cv=3,
    verbose=1,
    n_jobs=-1,
    random_state=42
)

random_search.fit(X_train, y_train)
```

# **12. Best Model**

```python
best_model = random_search.best_estimator_
print("Best Params:", random_search.best_params_)
```

# **13. Evaluation after tuning**


```python
y_pred = best_model.predict(X_val)
print("Tuned F1:", f1_score(y_val, y_pred))
```

# **14. Tuning with threshold for prediction**

```python
import numpy as np
y_prob = best_model.predict_proba(X_val)[:, 1]

best_f1 = 0
best_thresh = 0

for t in np.arange(0.2, 0.6, 0.05):
    preds = (y_prob > t).astype(int)
    score = f1_score(y_val, preds)

    if score > best_f1:
        best_f1 = score
        best_thresh = t

print("Best Threshold:", best_thresh)
print("Best F1:", best_f1)
```

# **15. Training on Full Data**

```python
best_model.fit(X, y)
```

# **16. Prediction on Test Data**

```python
test_prob = best_model.predict_proba(test_df)[:, 1]
final_preds = (test_prob > best_thresh).astype(int)
```

# **17. Create Submission File**

```python
submission = pd.DataFrame({
    'employee_id': sample_df['employee_id'],
    'is_promoted': final_preds
})
```

# **18. Save as csv**

```python
submission.to_csv("submission.csv", index=False)
submission.head(4)
```
