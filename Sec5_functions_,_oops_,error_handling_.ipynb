# List â€“ Python Example

## Class
**Definition:** A blueprint or template that defines what something is and can do.  
**Example:** `list` is a class â€“ it tells Python how lists behave.

## Object
**Definition:** A real instance of a class.  
**Example:** `fruits = ["apple", "banana", "cherry"]` â†’ `fruits` is a list object.

## Attributes
**Definition:** Properties or characteristics of a list object.  
**Example:**  
- `fruits[0]` â†’ `"apple"` (first element)  
- `fruits[1]` â†’ `"banana"` (second element)  
- `fruits[-1]` â†’ `"cherry"` (last element)

## Methods
**Definition:** Actions a list object can perform.  
**Example:**  
- `fruits.append("orange")` â†’ adds `"orange"` to the list  
- `fruits.remove("banana")` â†’ removes `"banana"` from the list  
- `fruits.reverse()` â†’ reverses the list

```python
# Class: Dog
# A class is a blueprint that defines what a Dog is and what it has
class Dog:

    # Method: __init__
    # This is a special method (constructor) that runs when an object is created
    # It initializes (sets) the attributes of the object
    def __init__(self, name, age):
        # Attributes of the Dog object
        self.name = name   # name attribute
        self.age = age     # age attribute

# Object: dog1
# dog1 is an object (instance) of the Dog class
dog1 = Dog("Buddy", 3)

# Accessing attributes of the object
print(dog1.name)  # Attribute access â†’ Buddy
print(dog1.age)   # Attribute access â†’ 3
```

# Create a class named Student.

Each student should have the following attributes:

name

roll_no

marks

Create a method called display() that prints:

```python
class Student:
  def __init__(self,name,roll_no,marks):
    self.name = name
    self.roll_no = roll_no
    self.marks = marks

  def display(self):
    print(self.name,[self.roll_no],'has scored',self.marks)

stu1 = Student("rani",6543,88)
stu1.display()
```
